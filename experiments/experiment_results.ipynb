```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('../pickles/runtimes.pkl', 'rb') as f:
    runtime_data = pickle.load(f)
```

```python
df = pd.DataFrame({
    'LoConsensus': runtime_data['loconsensus_runtimes'],
    'LocoMotif': runtime_data['locomotif_runtimes'],
    'F1 Scores': runtime_data['f1_scores']
})
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
plt.figure(figsize=(12, 6))
plt.boxplot([df['LoConsensus'], df['LocoMotif']], tick_labels=['LoConsensus', 'LocoMotif'])
plt.ylabel('Runtime (seconden)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(df['LoConsensus'], bins=30, density=True, alpha=0.7)
density = plt.gca().twinx()
x_range = np.linspace(df['LoConsensus'].min(), df['LoConsensus'].max(), 100)
kde = plt.mlab.GaussianKDE(df['LoConsensus'])
density.plot(x_range, kde(x_range), 'r-')
plt.xlabel('Runtime (seconden)')
plt.ylabel('Frequentie')
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 2)
plt.hist(df['LocoMotif'], bins=30, density=True, alpha=0.7)
density = plt.gca().twinx()
x_range = np.linspace(df['LocoMotif'].min(), df['LocoMotif'].max(), 100)
kde = plt.mlab.GaussianKDE(df['LocoMotif'])
density.plot(x_range, kde(x_range), 'r-')
plt.xlabel('Runtime (seconden)')
plt.ylabel('Frequentie')
```

```python
plt.figure(figsize=(8, 6))
plt.hist(runtime_data['f1_scores'], bins=30, density=True, alpha=0.7)
density = plt.gca().twinx()
x_range = np.linspace(min(runtime_data['f1_scores']), max(runtime_data['f1_scores']), 100)
kde = plt.mlab.GaussianKDE(runtime_data['f1_scores'])
density.plot(x_range, kde(x_range), 'r-')
plt.xlabel('F1 Score')
plt.ylabel('Frequentie')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
